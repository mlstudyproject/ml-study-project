Regression이란 data set을 참조하여 특정 결과가 어떻게 나올 수 있을지에 대한 학습된 data를 기반으로 결과값을 추정하는 것

Regression을 학습한다는 것은 어떤 가설(Hypothesis)을 세운다.
Linear Regression : data set이 Linear하다는 것을 가설로 세우고, 어떤 값들이 갖고 있는 data에 근접하는지를 판단하는 것이 학습하다는 의미
Hypothesis : H(x) = Wx + b 이 Linear한 식에서 W값, b값을 찾는 것이 된다.
Cost function : (H(x) - y)^2 우리가 세운 가설과 실제의 데이터가 얼마나 다른지를 나타내는 것. 이 값이 적은 것을 찾는 것이 학습이 되는 것.
=> minimize cost(W,b)

```python
import tensorflow as tf

x_data = [1, 2, 3]
y_data = [1, 2, 3]

W = tf.Variable(tf.random_uniform([1], -1.0, 1.0))
b = tf.Variable(tf.random_uniform([1], -1.0, 1.0))

#Our hypothesis
hypothesis = W * x_data + b

#Simplified cost function
cost = tf.reduce_mean(tf.square(hypothesis - y_data))

#Minimize
a = tf.Variable(0.1) #Learnig rate, alpha
optimizer = tf.train.GradientDescentOptimizer(a)
train = optimizer.minimize(cost)

# Before staring, initialize the variables. We will 'run' this first.
init = tf.global_variables_initializer()

sess = tf.Session()
sess.run(init)

for step in range(2001):
    sess.run(train)
    if step % 20 == 0:
        print(step, sess.run(cost), sess.run(W), sess.run(b))
```

코드 재사용을 위한 Placeholder를 사용한 예제

```python
import tensorflow as tf

x_data = [1, 2, 3]
y_data = [1, 2, 3]

W = tf.Variable(tf.random_uniform([1], -1.0, 1.0))
b = tf.Variable(tf.random_uniform([1], -1.0, 1.0))

#placeholder X,Y 값 설정
X = tf.placeholder(tf.float32)
Y = tf.placeholder(tf.float32)

#Our hypothesis
hypothesis = W * X + b

#Simplified cost function
cost = tf.reduce_mean(tf.square(hypothesis - Y))

#Minimize
a = tf.Variable(0.1) #Learnig rate, alpha
optimizer = tf.train.GradientDescentOptimizer(a)
train = optimizer.minimize(cost)

# Before staring, initialize the variables. We will 'run' this first.
init = tf.global_variables_initializer()

sess = tf.Session()
sess.run(init)

for step in range(2001):
    sess.run(train, feed_dict={X:x_data, Y:y_data})
    if step % 20 == 0:
        print(step, sess.run(cost, feed_dict={X:x_data, Y:y_data}), sess.run(W), sess.run(b))
        
print(sess.run(hypothesis, feed_dict={X:5}))
print(sess.run(hypothesis, feed_dict={X:2.5}))
```
